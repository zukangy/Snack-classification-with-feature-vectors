```python
import numpy as np 
import matplotlib.pyplot as plt
from utils.edge_detection import EdgeDetection
from utils.snack_dataset import SnackDataset
```

```python
target_imgs = {"Waffle": 6,
               "Apple": 15,
               "Hot dog": 7,
               "Popcorn": 3,
               "Ice cream": 17}
```

```python
def plot_feature_vector(img, feature):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.xticks(ticks=[])
    plt.yticks(ticks=[])

    plt.subplot(1, 2, 2)
    plt.imshow(feature, cmap='gray')
    plt.xticks(ticks=[])
    plt.yticks(ticks=[])
    plt.show()
```

```python
snacks = SnackDataset()
```

```python
apple_label = snacks.reverse_label_mapping('Apple')
images = list(snacks.get_images_by_label(apple_label))
img, label = images[target_imgs['Apple']]

edge = EdgeDetection(img)
canny_image = edge.canny()

plot_feature_vector(img, canny_image)
```
