```python
import matplotlib.pyplot as plt
from utils.create_hop import HOG
from utils.snack_dataset import SnackDataset
```

```python
snacks = SnackDataset()
```

```python
target_imgs = {"Waffle": 6,
               "Apple": 15,
               "Hot dog": 7,
               "Popcorn": 3,
               "Ice cream": 17}
```

```python
snacks = SnackDataset()
```

```python
def plot_feature_vector(img, feature):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.xticks(ticks=[])
    plt.yticks(ticks=[])

    plt.subplot(1, 2, 2)
    plt.imshow(feature, cmap='gray')
    plt.xticks(ticks=[])
    plt.yticks(ticks=[])
    plt.show()
```

```python
waffle_label = snacks.reverse_label_mapping('Apple')
images = list(snacks.get_images_by_label(waffle_label))
img, label = images[target_imgs['Apple']]

hog = HOG(img)
hog_image = hog.hog_image
hog_image_rescaled = hog.hog_image_rescaled

plot_feature_vector(img, hog_image)
```
